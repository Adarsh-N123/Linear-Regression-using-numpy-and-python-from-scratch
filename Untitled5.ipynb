```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams["figure.figsize"]=(12,8)
```

```python
df = pd.read_csv("bike_sharing_data.txt")
```

```python
sns.scatterplot(x="Population",y="Profit",data=df)
```

```python
def cost_func(x,y,theta):
  m = len(y)
  y_pred = x.dot(theta)
  error = (y_pred-y)**2
  return (1/(2*m))*(np.sum(error))
```

```python
m = df.Population.values.size
x = np.append(np.ones((m,1)),df.Population.values.reshape(m,1),axis=1)
y = df.Profit.values.reshape(m,1)
theta = np.zeros((2,1))

cost_func(x,y,theta)
```

```python
def gradient(x,y,theta,alpha,iterations):
  m = len(y)
  costs=[]
  for i in range(iterations):
    y_pred = x.dot(theta)
    error = np.dot(x.transpose(),(y_pred-y))
    theta -= alpha*1/m*error
    costs.append(cost_func(x,y,theta))
  return theta,costs
```

```python
theta,costs = gradient(x,y,theta,alpha = 0.01,iterations=2000)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
theta_0 = np.linspace(-20,20,100)
theta_1 = np.linspace(-2,8,100)
cost_vals = np.zeros((len(theta_0),len(theta_1)))
```

```python
for i in range(len(theta_0)):
  for j in range(len(theta_1)):
    t = np.array([theta_0[i],theta_1[j]])
    cost_vals[i,j] = cost_func(x,y,t)
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection="3d")
surf = ax.plot_surface(theta_0,theta_1,cost_vals,cmap='viridis')
fig.colorbar(surf,shrink=0.5,aspect=5)
plt.xlabel("$\Theta_0$")
plt.ylabel("$\Theta_1$")
ax.set_zlabel("J($\Theta$)")
plt.show()
```
